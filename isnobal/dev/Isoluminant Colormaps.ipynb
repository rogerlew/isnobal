```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import Normalize
%matplotlib inline

import numpy as np
from colormath.color_objects import LabColor, sRGBColor, AdobeRGBColor
from colormath.color_conversions import convert_color
```

```python
rgb = AdobeRGBColor(255/255.,   0, 238/255.)
```

```python
rgb
```

```python
convert_color(rgb, LabColor, target_illuminant='d65')
```

```python
lab_l = 75
avals = np.linspace(-40, 0, 128)
bvals = np.linspace(-40, 0, 128)
```

```python
avals = np.concatenate((list(avals)[:-1], np.linspace(0, 70, 129)))
bvals = np.concatenate((list(bvals)[:-1], np.linspace(0,-30, 129)))
len(avals), len(bvals)
```

```python
C = []
LAB = []
for a, b in zip(avals, bvals):
    lab_c = LabColor(lab_l, a, b, illuminant='d65')
    c = convert_color(lab_c, AdobeRGBColor)
    C.append((c.rgb_r, c.rgb_g, c.rgb_b, 1.0))
    LAB.append((lab_c.lab_l, lab_c.lab_a, lab_c.lab_b))
```

```python
len(C)
```

```python
def _buildLinearSegmentedColormap(x, C, name):
    cdict = dict(red=[], green=[], blue=[], alpha=[])

    for j, boundary in enumerate(x):
        for i, c in enumerate('red green blue alpha'.split()):
            y0 = y1 = C[j][i]
            cdict[c].append((boundary, y0, y1))
            
    return  mpl.colors.LinearSegmentedColormap(name, cdict)
```

```python
_delta_Q = np.array(C, dtype=np.float32)
_delta_Q_x = np.linspace(0, 1, 256)

delta_Q = _buildLinearSegmentedColormap(_delta_Q_x, _delta_Q, 'delta_Q')

data = [[j for j in xrange(1001)] for i in xrange(100)]
data = np.array(data, dtype=np.float32)
data /= 1000.0

plt.figure(figsize=(20,8))
plt.subplot(2,1,1)
plt.imshow(delta_Q(data))

plt_cool_cm = plt.cm.cool
plt.subplot(2,1,2)
plt.imshow(plt_cool_cm(data))
```
